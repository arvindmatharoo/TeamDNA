```python
import requests
import pandas as pd

# Assuming comp_id and season_id are already set
url_matches = f"https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/{comp_id}/{season_id}.json"
matches = requests.get(url_matches).json()
df_matches = pd.DataFrame(matches)

# Extract nested fields
df_matches['home_team_name'] = df_matches['home_team'].apply(lambda x: x['home_team_name'])
df_matches['away_team_name'] = df_matches['away_team'].apply(lambda x: x['away_team_name'])
df_matches['stage'] = df_matches['competition_stage'].apply(lambda x: x['name'])
df_matches['match_label'] = df_matches['home_team_name'] + " vs " + df_matches['away_team_name']

def get_knockout_winner(row):
    knockout_stages = ['Round of 16','Quarter-finals','Semi-finals','Final']

    if row['stage'] in knockout_stages :
        home = row.get('home_score_extra_time',row['home_score'])
        away = row.get('away_score_extra_time',row['away_score'])
        if home>away:
            return row['home_team_name']
        elif away > home :
            return row['away_team_name']
        else :
            #Penalties 
            home_pen = row.get('home_score_penalties',0)
            away_pen = row.get('away_score_penalties',0)
            if home_pen is not None and away_pen is not None :
                if home_pen > away_pen:
                    return row['home_team_name']
                elif away_pen > home_pen :
                    return row['away_team_name']
                else:
                    return "Draw"
            else :
                return "Draw"
    else :
        if row['home_score'] > row['away_score']:
            return row['home_team_name']
        elif row['home_score'] < row['away_score']:
            return row['away_team_name']
        else : 
            return "Draw"

df_matches['winner'] = df_matches.apply(get_knockout_winner,axis=1)

# Display only relevant columns
df_display = df_matches[['match_id', 'stage', 'match_label', 'home_score', 'away_score','winner', 'match_date']]
df_display= df_display.rename(columns={
    'home_score': 'home_goals',
    'away_score': 'away_goals',
    'match_date': 'date'
})

print("\n=== Matches in Selected Competition/Season ===")
print(df_display.to_string(index=False))
```

```python
teams = pd.unique(df_matches[['home_team_name','away_team_name']].values.ravel())

team_stats = []

for team in teams :
    home = df_matches[df_matches['home_team_name']==team]
    away = df_matches[df_matches['away_team_name']==team]

    matches_played = len(home)+len(away)
    wins = sum(home['winner'] == team) + sum(away['winner'] == team)
    draws = sum(home['winner'] == "Draw") + sum(away['winner'] == "Draw")
    losses = matches_played-wins-draws 
    goals_scored = home['home_score'].sum() + away['away_score'].sum()
    goals_conceded = home['away_score'].sum() + away['home_score'].sum()
    goals_diff =  goals_scored - goals_conceded

    team_stats.append({
    'team': team,
    'matches_played':matches_played,
    'wins':wins,
    'draws':draws,
    'losses':losses,
    'goals_scored':goals_scored,
    'goals_conceded':goals_conceded,
    'goal_difference':goals_diff
    })
df_team_stats = pd.DataFrame(team_stats)

print("\n====TEAM LEVEL SUMMARY STATS ====")
print(df_team_stats.sort_values('matches_played',ascending=False).to_string(index=False))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# --- 1. Filter only passes ---
df_passes = df_events[df_events['type'].apply(
    lambda x: x['name'] if isinstance(x, dict) else None
) == "Pass"].copy()

# --- 2. Extract passer, receiver, and coordinates ---
df_passes.loc[:, 'passer'] = df_passes['player'].apply(
    lambda x: x['name'] if isinstance(x, dict) else None
)
df_passes.loc[:, 'receiver'] = df_passes['pass'].apply(
    lambda x: x['recipient']['name'] if isinstance(x, dict) and x.get('recipient') else None
)
df_passes.loc[:, 'x'] = df_passes['location_x']
df_passes.loc[:, 'y'] = df_passes['location_y']

# --- 3. Build directed weighted graph (edges = passes) ---
G = nx.DiGraph()
for _, row in df_passes.dropna(subset=['passer','receiver']).iterrows():
    if G.has_edge(row['passer'], row['receiver']):
        G[row['passer']][row['receiver']]['weight'] += 1
    else:
        G.add_edge(row['passer'], row['receiver'], weight=1)

# --- 4. Compute average positions for players ---
player_positions = df_passes.groupby('passer')[['x','y']].mean().reset_index()
pos = {row['passer']: (row['x'], 80-row['y']) for _, row in player_positions.iterrows()}  
# Flip y-axis for correct pitch orientation

# --- 5. Draw pitch + passing network ---
plt.figure(figsize=(10,7))
ax = plt.gca()
draw_pitch(ax)

# Draw players (nodes)
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color="skyblue", ax=ax)
nx.draw_networkx_labels(G, pos, font_size=9, ax=ax)

# Draw passes (edges with weight scaling)
nx.draw_networkx_edges(
    G, pos,
    arrowstyle='->', arrowsize=15,
    edge_color="red",
    width=[d['weight']*0.5 for _,_,d in G.edges(data=True)],
    ax=ax
)

plt.title("Team Passing Network (Position-based)", color="white")
plt.show()
```

```python
df_shots = df_events[df_events['type'].apply(lambda x:x['name'] if isinstance(x,dict) else None) == "Shot"].copy()


df_shots.loc[:,'team'] = df_shots['team'].apply(lambda x:x['name'] if isinstance(x,dict) else None)
df_shots.loc[:,'player_name'] = df_shots['player'].apply(lambda x:x['name'] if isinstance(x,dict) else None)
df_shots.loc[:,'x'] = df_shots['location_x']
df_shots.loc[:,'y'] =df_shots['location_y']
df_shots.loc[:,'outcome'] = df_shots['shot'].apply(lambda x:x['outcome']['name'] if isinstance(x,dict) and x.get('outcome') else None)
df_shots.loc[:,'xg'] = df_shots['shot'].apply(lambda x:x['statsbomb_xg'] if isinstance(x,dict) and x.get('statsbomb_xg') else 0)


teams = df_shots['team'].unique()

fig,axes = plt.subplots(1,len(teams),figsize=(18,7), facecolor = 'black')

if len(teams) == 1 :
    axes =[axes]

for i, team in enumerate(teams):
    ax = axes[i]
    draw_pitch(ax)

    df_team = df_shots[df_shots['team'] == team]

    non_goals = df_team[df_team['outcome'] != 'Goal']
    ax.scatter(non_goals['x'], 80 - non_goals['y'],
    s = non_goals['xg']* 800, c ='skyblue',edgecolors ='white',alpha=0.6,label="Non-Goals")

    goals = df_team[df_team['outcome'] == 'Goal']
    ax.scatter(goals['x'],80-goals['y'],s=goals['xg']* 1200,
    c='red',marker='*',edgecolors='white',linewidth =1.2, label="Goals")

    ax.set_title(f"{team} Shot Map (xG)",color='white')
plt.show()
```

```python
# Corrected xG Summary Code

xg_summary = (
    df_shots.groupby(['team', 'player_name'])
    .agg(
        Shots=('xg', 'count'),
        Total_xG=('xg', 'sum'),
        Goals=('outcome', lambda x: (x == "Goal").sum())
    )
    .reset_index()
)

# Add Goals per xG ratio
xg_summary['Goals_per_xG'] = (xg_summary['Goals'] / xg_summary['Total_xG']).round(2)

# Sort
xg_summary = xg_summary.sort_values(['team', 'Total_xG'], ascending=[True, False])

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=xg_summary, x='Total_xG', y='player_name', hue='team')
plt.title("Expected Goals (xG) Contribution per Player", fontsize=14, color="black")
plt.xlabel("Total xG")
plt.ylabel("Player")
plt.legend(title="Team")
plt.show()

# Display summary table
xg_summary
```
